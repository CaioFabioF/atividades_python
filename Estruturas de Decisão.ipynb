Round()
Função que determina se o número é redondo ( inteiro ) ou decimal

```python
a = float(input("Digite um número: "))
b = float(input("Digite um número: "))
operação = input("Digite a operação que deseja realizar. 1 para decimal ou inteiro, 2 para par ou ímpar, 3 para negativo ou positvo: ")
if operação == "1":
  if a == round(a) and b == round(b):
    print("Os números são inteiros")
  elif a == round(a) and b != round(b):
    print("O primeiro número é inteiro e o segundo é decimal")
  elif a != round(a) and b == round(b):
    print("O primeiro número é decimal e o segundo é inteiro")
  else:
    print("Os números são decimais")
elif operação == "2":
  if a % 2 == 0 and b % 2 == 0:
    print("Os números são pares")
  elif a % 2 == 0 and b % 2 != 0:
    print("O primeiro número é par e o segundo é ímpar")
  elif a % 2 != 0 and b % 2 == 0:
    print("O primeiro número é ímpar e o segundo é par")
  else:
    print("Os números são ímpares")
elif operação == "3":
  if a > 0 and b > 0:
    print("Os números são positivos")
  elif a > 0 and b < 0:
    print("O primeiro número é positivo e o segundo é negativo")
  elif a < 0 and b > 0:
    print("O primeiro número é negativo e o segundo é positivo")
  else:
    print("Os números são negativos")
else :
  print("Operação inválida")
```

```python
# Função para realizar as perguntas e contar as respostas "sim"
def interrogar():
    respostas_positivas = 0

    # Fazendo as 5 perguntas e contabilizando as respostas "sim"
    resposta = input("Telefonou para a vítima? ")
    if resposta == "sim":
        respostas_positivas += 1

    resposta = input("Esteve no local do crime? ")
    if resposta == "sim":
        respostas_positivas += 1

    resposta = input("Mora perto da vítima? ")
    if resposta == "sim":
        respostas_positivas += 1

    resposta = input("Devia para a vítima? ")
    if resposta == "sim":
        respostas_positivas += 1

    resposta = input("Já trabalhou com a vítima? ")
    if resposta == "sim":
        respostas_positivas += 1

    return respostas_positivas

# Função para classificar a participação com base nas respostas positivas
def classificar_participacao(respostas_positivas):
    if respostas_positivas == 2:
        return "Suspeita"
    else:
        if 3 <= respostas_positivas <= 4:
            return "Cúmplice"
        else:
            if respostas_positivas == 5:
                return "Assassino"
            else:
                return "Inocente"

# Executa a interrogação e classifica a participação
respostas_positivas = interrogar()
classificacao = classificar_participacao(respostas_positivas)

# Exibe a classificação final
print(f"A classificação é: {classificacao}")

```
